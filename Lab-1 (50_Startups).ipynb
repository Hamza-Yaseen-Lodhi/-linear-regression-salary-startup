```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn import metrics
```

```python
from sklearn.linear_model import LinearRegression
```

```python
sns.pairplot(df)
```

```python
X = df[['R&D Spend','Administration','Marketing Spend']]
y = df[['Profit']]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=43)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_predicted = model.predict(X_test)
```

```python
# let's evalutate whether our y_predicted is how much near the actual y value
from sklearn.metrics import  mean_squared_error,r2_score
```

```python
MSE = mean_squared_error(y_test,y_predicted)
print(f"The mean square error of model is {MSE}")
print(f"The root mean square error of model is {np.sqrt(MSE)}")
```

```python
performance=r2_score(y_test,y_predicted)
print(f"The r_2 score of the model is {performance}")
```

```python
# let's plot this to check how well is this working 
plt.figure(figsize=(8, 5))

# Plotting each feature in X_test against y_test and y_predicted
for i, feature in enumerate(X_test.columns):
	plt.scatter(X_test[feature], y_test, color='blue', label='Actual Profit', alpha=0.7)
	plt.scatter(X_test[feature], y_predicted, color='red', label='Predicted Profit', alpha=0.7)
	plt.plot(X_test[feature], y_predicted, color='green', label='Regression Line')
	plt.title(f'Actual vs Predicted Profit for {feature}')
	plt.xlabel(feature)
	plt.ylabel('Profit')
	plt.legend()
	plt.show()
```

```python
# save the model to a file
import joblib

# Assuming 'model' is your trained regression model
joblib.dump(model, 'Hamza.pkl')
```
